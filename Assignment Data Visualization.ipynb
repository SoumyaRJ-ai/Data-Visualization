### In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
fig = plt.figure(figsize=(8,5))
ax = Axes3D(fig)
X = decomposition.PCA().fit_transform(iris.data)
ax.scatter3D(X[:,0], X[:,1], X[:,2], c = iris.target, cmap = plt.cm.Set1,edgecolor='red', s = 60)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```
